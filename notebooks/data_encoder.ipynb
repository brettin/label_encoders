```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
pd.options.display.max_rows = 1000
pd.options.display.max_columns = 1000
```

```python
infile  = '../data/TRACKTBI_Pilot_DEID_02.22.18v2.2.dos.txt'
outfile = '../data/TRACKTBI_Pilot_DEID_02.22.18v2.2.labelencoded.csv'

df = pd.read_csv(infile, delimiter="\t")
df.head()
```

```python
print ('original shape: ', df.shape)
```

```python
# Drop the patient id
df = df.drop(columns=['PatientNum'])

# Replace cells with empty string or string of white space(s)
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# Label encoding: NOTE: cat.codes will put -1 in for NaNs.
# We will encode those columns that have only one or two
# values with 0 and 1. One trick you can use in pandas is
# to convert a column to a category, then use those category
# values for your label encoding.
for header in df:
    if (df[header].dtype == 'object'):
        s = df[header].value_counts()
        if (s.size == 1 or s.size == 2):
            df[header] = df[header].astype('category')
            df[header] = df[header].cat.codes
            
df.head()
```

```python
# Next let's one hot encode columns with 3-5 values.
new = 0
removed = 0
for header in df:
    if (df[header].dtype == 'object'):
        s = df[header].value_counts()
        if (2 < s.size <= 5):
            new = new + s.size
            removed = removed + 1
            df = pd.get_dummies(df, columns=[header])
print ('total columns created: ', new)
print ('total columns removed: ', removed)
print ('shape after one hot encoding: ', df.shape)
```

```python
df.head(n=10)
```

```python
# count the columns left for binary encoding
o_count = 0
for header in df:
    if (df[header].dtype == 'object'):
        o_count = o_count + 1
print ('objects left in dataframe: ',o_count)
```

```python
from category_encoders import *

# NOTE: cat.codes will put -1 in for NaNs. Need to verify
# what that looks like after binary encoder
be_cols=[]
for header in df:
    if (df[header].dtype == 'object'):
        s = df[header].value_counts()
        if (s.size > 5):
            df[header] = df[header].astype('category')
            df[header] = df[header].cat.codes
            be_cols.append(header)
be = BinaryEncoder(cols=be_cols).fit(df)
df = be.transform(df)
print ('shape after binary encoding: ', df.shape)
```
